#### **Introduction**: This report presents an in-depth analysis and implementation of "AdaptKeyBERT," an innovative approach for Few-Shot & Zero-Shot Domain Adaptation of KeyBERT, focusing on keyword extraction in natural language processing.

#### **Problem Description**:

Keyword extraction is crucial in NLP but faces challenges in limited data scenarios. Existing models often require extensive labeled data, making them less effective in low-resource domains.

#### **Context of the Problem**:

Efficient keyword extraction is essential in various fields like biomedicine, legal document processing, and information retrieval, especially when dealing with specialized or low-resource domains.

#### **Limitation About other Approaches**:

Previous approaches, primarily relying on large datasets for training, struggled with domain adaptation in scenarios with scarce or specific data.

#### **Solution**:

AdaptKeyBERT addresses this by incorporating Few-Shot and Zero-Shot Learning with an attention-based mechanism into KeyBERT, enhancing domain-specific keyword extraction with limited data.

```python
import gradio as gr
from adaptkeybert import KeyBERT

def extract_keywords(doc, top_n=10):
    kw_model = KeyBERT()
    keywords = kw_model.extract_keywords(doc, top_n=top_n)
    return keywords

interface = gr.Interface(
    fn=extract_keywords,
    inputs=[
        gr.Textbox(lines=10, placeholder="Enter text here..."),
        gr.Slider(minimum=1, maximum=20, value=10, label="Top N")
    ],
    outputs="text",
    title="Keyword Extraction with KeyBERT",
    description="Enter text to extract keywords using KeyBERT."
)

interface.launch()
```

```python
def extract_keywords(doc, domain_adapt, pre_train_examples, lr, top_n):
    kw_model = KeyBERT(domain_adapt=domain_adapt)
    if pre_train_examples:
        kw_model.pre_train([doc], pre_train_examples, lr=lr)
    keywords = kw_model.extract_keywords(doc, top_n=top_n)
    return keywords

iface = gr.Interface(
    fn=extract_keywords,
    inputs=[
        gr.Textbox(label="Document Text", placeholder="Enter document text here..."),
        gr.Checkbox(label="Domain Adaptation", value=True),
        gr.Textbox(label="Pre-Train Examples (as a list of lists)", placeholder="[['supervised', 'unsupervised']]"),
        gr.Number(label="Learning Rate", value=1e-3),
        gr.Slider(label="Top N Keywords", minimum=1, maximum=20, value=10)
    ],
    outputs="text",
    title="Keyword Extraction with KeyBERT",
    description="Enter text to extract keywords using KeyBERT."
)

iface.launch()
```

```python
from adaptkeybert import KeyBERT

doc = """
         Supervised learning is the machine learning task of learning a function that
         maps an input to an output based on example input-output pairs. It infers a
         function from labeled training data consisting of a set of training examples.
         In supervised learning, each example is a pair consisting of an input object
         (typically a vector) and a desired output value (also called the supervisory signal).
         A supervised learning algorithm analyzes the training data and produces an inferred function,
         which can be used for mapping new examples. An optimal scenario will allow for the
         algorithm to correctly determine the class labels for unseen instances. This requires
         the learning algorithm to generalize from the training data to unseen situations in a
         'reasonable' way (see inductive bias). But then what about supervision and unsupervision, what happens to unsupervised learning.
      """
kw_model = KeyBERT()
keywords = kw_model.extract_keywords(doc, top_n=10)
print(keywords)
```

```python
kw_model = KeyBERT(domain_adapt=True)
kw_model.pre_train([doc], [['supervised', 'unsupervised']], lr=1e-3)
keywords = kw_model.extract_keywords(doc, top_n=10)
print(keywords)
```

```python
kw_model = KeyBERT(domain_adapt=True, zero_adapt=True)
kw_model.pre_train([doc], [['supervised', 'unsupervised']], lr=1e-3)
kw_model.zeroshot_pre_train(['supervised', 'unsupervised'], adaptive_thr=0.15)
keywords = kw_model.extract_keywords(doc, top_n=10)
print(keywords)
```
